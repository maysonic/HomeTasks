На гистограмме имеются явные пики. Будем считать границами ценовых групп средние значения между двумя соседними пиками высотой >= 40.

"Список" пиков высотой >= 40:

```python
price_peaks = (
    data.query('last_price < @last_price_border')
    .groupby('last_price')
    ['last_price'].count()
)
price_peaks_40 = price_peaks[price_peaks>=40]
```

Создадим список c границами ценовых групп:

```python
price_borders_list = []
for price_index in range(0,len(price_peaks_40)-1):
    #Добавляем к списку среднее между двумя соседними пиками:
    price_borders_list.append(
        (price_peaks_40.index[price_index] + price_peaks_40.index[price_index+1]) / 2
    )
#Добавляем к списку последнюю границу и максимальную цену в столбце:
price_borders_list.append(last_price_border)
price_borders_list.append(data['last_price'].max())
```

Всего пулучилось групп:

```python
len(price_borders_list)
```

Объявим функцию определения ценовой группы:

```python
def add_price_group(row):
    compared_price = row['last_price']
    price_group = 1
    #Если полученное в функцию значение меньше значения в отсортированном списке ценовых границ, 
    #то вернуть значение счетчика
    for price_border in price_borders_list:
        if compared_price <= price_border:
            return price_group
        price_group += 1
    return price_group
```

Создадим столбец с ценовыми группами рядом со столбцом **last_price**:

```python
try:
    data.insert(loc=2, column='price_group', value='')
except:
    pass
```

Заполним новый столбец **price_group**:

```python
data['price_group'] = data.apply(add_price_group, axis=1)
```

Данные о наполненности каждой ценовой группы:

```python
data.groupby('price_group')['price_group'].count().hist(bins=len(price_borders_list))
```

#### А. Изучение пропусков в столбце

В столбце 9195 пропущенных значений. Построим гистограмму для объявлений с такими пропусками. Сгруппируем строки с отсутствующим признаком **ceiling_height** по датам и посчитаем сколько объявлений с пропуском приходится на каждую дату:

В качестве дополнительной оценки случайности посчитаем коэффициент корреляции дат появления объявлений с пропуском с датами подачи всех объявлений. Объявим функцию, которая пригодится позже:

```python
def calculate_date_corr(column_count):
    corr = (
        data.groupby('first_day_exposition')['first_day_exposition'].count().
        corr(column_count)
    )
    return corr
```

```python
areas_by_price_groups = data.pivot_table(index='price_group', 
                                         values=['total_area', 'living_area', 'kitchen_area'], 
                                         aggfunc='median')
```

```python
areas_by_price_groups.head()
```

Объявим функцию для заполнения пропусков в столбце **living_area**. 
Возвращаемое значение равно: **total_area** * **K**,  
где **К** - отношение медиан жилой и общей площадей для заданной ценовой группы

```python
def living_area_fillna(row):
    price_group = row['price_group']
    new_living_area = (
        row['total_area'] 
        * areas_by_price_groups.loc[price_group, 'living_area'] 
        / areas_by_price_groups.loc[price_group, 'total_area']
    )
    return round(new_living_area, 1)
```

Заполним пропуски:

Объявим функцию, возвращающую границы нормальных значений признака. Границами нормальных значений примем межквантильные отступы\*1.5 от 25 и 75-го квантилей. 

```python
def get_type_values(column, data):
    IQR_mustache = (data.loc['75%',column] - data.loc['25%',column]) * 1.5
    data_min = data.loc['min',column]
    data_max = data.loc['max',column]
    calc_min = data.loc['25%',column] - IQR_mustache
    calc_max = data.loc['75%',column] + IQR_mustache
    if calc_min < data_min:
        calc_min = data_min
    if calc_max > data_max:
        calc_max = data_max
    return round(calc_min,1), round(calc_max,1)
```
